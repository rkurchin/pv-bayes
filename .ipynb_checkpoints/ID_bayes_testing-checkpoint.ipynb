# The Old Way

In the Bayes paper on SnS, we fitted four parameters ($\mu_e$ and $\tau_e$ in the SnS layer, and SRV and $\Delta E_c$ at the SnS-Zn(O,S) interface). This four-dimensional parameter space was sampled on a grid that was 16-20 boxes on a side. Accounting for six observation conditions (three temperatures and two light intensities), this meant we had to run over *half a million* SCAPS simulations. Not fun. And imagine if we'd wanted to fit another parameter - that takes that number up by an order of magnitude!

Let's do some estimation - suppose we now measure at ten conditions and each parameter is divided into ten chunks. Then, as a function of the number of parameters to fit, the number of simulations to run looks like:

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

num_params = range(1, 7, 1) # up to six parameters
num_sims = 10*np.power(num_params, 10) # just exponentiate each element in that list
plt.plot(num_params, num_sims)
plt.xlabel('# parameters to fit')
plt.ylabel('# sims to be run')
plt.show()
```

Now suppose we want to increase the precision (i.e. divide the parameter range into smaller chunks). What happens then?

```python
num_divisions = range(10,30,5)
for num in num_divisions:
    plt.semilogy(num_params, 10*np.power(num_params,num))
plt.xlabel('# parameters to fit')
plt.ylabel('# sims to run')
plt.legend(('10 divisions','15 divisions','20 divisions','25 divisions'))
plt.show()
```

Let's all just appreciate that that's a semilog plot. Doubling the desired precision (i.e. going from 10 to 20 divisions) at fitting just four parameters requires going from about 10 million simulations to over 10 **quadrillion**.

Hopefully I've convinced you that this is not a tenable situation if we want larger numbers of parameters (as we'll need for more "wild-type" materials and reasonable precision in fitting (and/or just large uncertainty about the range over which to try). So with that, I bring you...

# The New Way!

I wrote some new code that does this better! Namely, the `Pmf` and `Param_point` Python classes, the former of which is inherited by `ID_Pmf` - a version to test it out using ideal diode fits, and `ID_bayes` which implements the relevant likelihood function and some visualization stuff for this test case.

All of this is on Github [here](https://github.com/rkurchin/pv-bayes/tree/point_objects) - at the moment in a public repository because my SSH keys for the private MIT Github are broken and I haven't figured out how to fix them yet.

```python
import ID_bayes as bayes
import ID_pmf as pmf
```

## Initialize some stuff

Our "measurements" will be on a diode with $n=1.41$ and $J_0=6$. We'll start with five subdivisions in $n$ (linearly spaced between 1 and 2) and ten in $J_0$ (logarithmically spaced between 0.1 and 100). "Measurements" at three temperatures and from -0.1 V to 0.2 V.

```python
test = pmf.ID_Pmf([5,10])
voltages = np.arange(-0.1, 0.2, 0.05)
currents_100 = [test.compute_ID(V,100,{'n':1.41,'J_0':6}) for V in voltages]
currents_300 = [test.compute_ID(V,300,{'n':1.41,'J_0':6}) for V in voltages]
currents_500 = [test.compute_ID(V,500,{'n':1.41,'J_0':6}) for V in voltages]

num_sims = [len(test.points)]
```

Before we jump into the Bayesian updating, let's look at the structure of the `Pmf` object itself:

```python
print test
```

We can pick out one of those points to look at as well.

```python
print test.points[17]
```

## First Bayes

Let's feed in a point on the JV curve at 300 K.

```python
voltages, currents_300
```

Let's make up some pretty large errors for the purposes of demonstration.

```python
J_err_300 = [2,1,0.5,5,20,100,500]
```

Now we'll calculate the likelihood as a result of feeding in the point at 0.1 V. The likelihood function assumes a Gaussian error about the simulated point with a standard deviation given by what I've here called `J_err_300`.

```python
lkl = bayes.likelihood(test, currents_300[4], voltages[4],300, J_err_300[4])
```

Now, we could just look at the numbers for the probabilities, even choosing the highest-probability spots...

```python
for point in lkl.most_probable(10):print point
```

But that's awkward and unillustrative, so it'd be nice to visualize. Fortunately, I wrote a function for that! It can do it two different ways! The first uses kernel density estimation, or KDE - read more about that [here](http://scikit-learn.org/stable/modules/density.html). The other is just a hexbin histogram.

Worth noting - to make the KDE looks nice (and the visualization generally more intuitive), I'm resampling the probabilities and drawing from a uniform distribution across each "box," even though the simulations are strictly run only at the center. As the boxes get smaller, this distinction won't matter.

```python
%matplotlib inline
    
bayes.visualize_probs(lkl, 'kde')
```

```python
bayes.visualize_probs(lkl, 'hex')
```

## Onward!

Okay, now let's keep going! We'll do some Bayesian updating and subdivisions "by hand" and visualize the PMF at each step along the way.

```python
test.multiply(lkl)
```

That's the update (note that the `multiply` function also renormalizes the PMF afterwards)! Now let's subdivide all cells with a probability higher than 0.1.

```python
test.subdivide(0.01)
num_sims.append(len(test.points))
```

```python
print test
```

Check the distribution still looks the same after the subdivision:

```python
bayes.visualize_probs(test, 'kde')
```

```python
for point in test.most_probable(10):print point
```
